```python
# Initialize two lists for storing ranges
section_assignment_1 = []
section_assignment_2 = []

# Read the file
with open('input_day4.txt', 'r') as file:
    for line in file:
        # Split each line into two parts
        range1, range2 = line.strip().split(',')

        # Convert each part into a range or tuple
        start1, end1 = map(int, range1.split('-'))
        start2, end2 = map(int, range2.split('-'))

        # Append to respective lists
        section_assignment_1.append((start1, end1))
        section_assignment_2.append((start2, end2))

# Output the lists
print("section_assignment_1:", section_assignment_1)
print("section_assignment_2:", section_assignment_2)
```

```python
def check_containment(range1, range2):
    """
    Checks if range1 is completely contained within range2 and vice versa.
    Both range1 and range2 are expected to be tuples (start, end).
    Returns a tuple of two booleans (range1_in_range2, range2_in_range1).
    """
    start1, end1 = range1
    start2, end2 = range2
    range1_in_range2 = start2 <= start1 and end1 <= end2
    range2_in_range1 = start1 <= start2 and end2 <= end1
    return range1_in_range2, range2_in_range1

# Counter for pairs where one range is contained within the other
containment_counter = 0

# Comparing each pair of ranges
for r1, r2 in zip(section_assignment_1, section_assignment_2):
    r1_in_r2, r2_in_r1 = check_containment(r1, r2)
    # Increment the counter if either containment check is True
    if r1_in_r2 or r2_in_r1:
        containment_counter += 1

# Output the total count of pairs with at least one containment
print(f"Total pairs with at least one containment: {containment_counter}")
```

```python
def check_overlap(range1, range2):
    """
    Checks if there is an overlap of at least one value between range1 and range2.
    Both range1 and range2 are expected to be tuples (start, end).
    Returns True if there is an overlap, otherwise False.
    """
    start1, end1 = range1
    start2, end2 = range2
    # Check for overlap
    return max(start1, start2) <= min(end1, end2)

# Assuming list1 and list2 are as defined in your scenario
list1 = [(2, 4), (2, 3), (5, 7), (2, 8), (6, 6), (2, 6)]
list2 = [(6, 8), (4, 5), (7, 9), (3, 7), (4, 6), (4, 8)]

# Counter for pairs where there is an overlap
overlap_counter = 0

# Comparing each pair of ranges for overlap
for r1, r2 in zip(section_assignment_1, section_assignment_2):
    if check_overlap(r1, r2):
        overlap_counter += 1

# Output the total count of pairs with overlap
print(f"Total pairs with overlap: {overlap_counter}")
```
